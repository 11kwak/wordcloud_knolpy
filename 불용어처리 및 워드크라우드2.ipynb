1. 자치법규목록_시도전체_사회적

```python
f = open('./data/자치법규목록_시도전체_사회적경제.txt','r',encoding="cp949")
txt = f.readlines()
f.close()
```

```python
new_txt = []
for i in txt:
    new_txt.append(i.replace("\n",""))
```

```python
str_txt = ""
for i in new_txt:
    str_txt = str_txt + i + ' '
```

```python
from collections import Counter
c = Counter(new_txt).most_common()
```

```python
print(len(c))
c
```

```python
stop = []

for i in range(267):
    stop.append(c[i][0])

# 필요한 키워드 살리기 
stop.remove("사회적경제")
stop.remove("지원")
stop.remove("육성")
stop.remove("구매촉진")
stop.remove("판로지원")
stop.remove("활성화")
stop.remove("사회적경제기업")
stop.remove("제품")
stop.remove("사회적경제제품")
stop.remove("시행규칙")
stop.remove("육성지원")
stop.remove("설치")
stop.remove("운영")
stop.remove("운용")
stop.remove("촉진")
stop.remove("사회적경제기금")
stop.remove("사회적경제기업제품")
stop.remove("구매")
stop.remove("사회적")
stop.remove("경제")
stop.remove("규칙")
stop.remove("판로개척")
stop.remove("기본조례")
stop.remove("사회적경제활동")
stop.remove("육성·지원")
stop.remove("사회적경제지원센터")
stop.remove("협의회")
stop.remove("구성")
stop.remove("교육")
stop.remove("기업")
stop.remove("제품구매")
stop.remove("육성ㆍ지원")
stop.remove("지원조례")
stop.remove("규정")
stop.remove("사회적경제활성화기금")
stop.remove("사회적가치")
stop.remove("실현")
stop.remove("마을")
stop.remove("마을만들기")
stop.remove("생산제품")
stop.remove("혁신타운")
stop.remove("조직")
stop.remove("사회적경제지원기금")



# stop = ['','.','에','의','위' ]

stopwords = set(stop)
```

```python
mask1 = np.array(Image.open("./data/수정됨22_동그라미.png"))

wc = WordCloud(background_color="white",font_path="C:\Windows\Fonts\H2GTRE.TTF",max_words=1000, mask=mask1, stopwords=stopwords, margin=10,
               random_state=1,contour_color='black').generate(str_txt)

wc.to_file("자치법규목록_시도전체_사회적경제_워드클라우드1차.png")
plt.axis("off")
plt.figure(figsize=(10,10))
plt.title("Default colors")
plt.imshow(wc, interpolation="bilinear")
plt.axis("off")
plt.show()
```
